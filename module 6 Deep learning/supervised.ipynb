```python
import pandas as pd
```

```python
df=pd.read_csv("price.csv")
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
model=LabelEncoder()
```

```python
df['city']=model.fit_transform(df['City'])
```

```python
df
```

```python
dummies=pd.get_dummies(df['City'])
dummies
```

```python
df['City']
df
```

```python
df_dummies=pd.concat([df,dummies],axis=1)
df_dummies
```

```python
df_final=df_dummies.drop('City',axis='columns')
df_final
```

```python
from sklearn.model_selection import train_test_split
```

```python
x=df_final.drop(['Price','Delhi','Kolkata','Mumbai'],axis='columns')
y=df_final['Price']
print(x)
print(y)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y)
```
